```python
import pandas as pd
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from nerodia.browser import Browser
import time
from selenium.webdriver.chrome.options import Options
import warnings
warnings.filterwarnings("ignore")
```

```python
options = Options()
options.add_argument('--headless')
options.add_argument('--no-sandbox') # Remove if running outside deepnotel
```

```python
def Pokemon():
    browser = Browser('chrome', options=options) # Create Browser
    browser.goto('https://www.serebii.net/pokemon/gen1pokemon.shtml') #LINK de acesso ao dataset
    # browser.goto('https://pokemondb.net/pokedex/all')
    time.sleep(4) # Coloda o scrip em pausa enquanto a página carrega

    df = pd.read_html(browser.html)[0]
    df = df.drop([1,3], axis=1)
    df = df.dropna()
    df = df.reset_index(drop=True)
    novo_cabecalho = df.iloc[1].tolist()
    df = df.iloc[2:]
    df.columns = novo_cabecalho
    df['No.'] = df['No.'].str[-3:]
    df.loc[df['Name'] == 'Nidoran♀', 'Name'] = 'Nidoran-femea'
    df.loc[df['Name'] == 'Nidoran♂', 'Name'] = 'Nidoran'
    df.loc[df['Name'] == "Farfetch'd", 'Name'] = 'farfetchd'
    df.loc[df['Name'] == "Mr. Mime", 'Name'] = 'mrmime'
    return df
```

```python
df = Pokemon()
```

```python
def types():
    browser = Browser('chrome', options=options) # Create Browser
    browser.goto('https://charizardpoke.blogspot.com/2014/08/pokedex-kanto1-geracao.html') #LINK de acesso ao dataset
    # browser.goto('https://pokemondb.net/pokedex/all')
    time.sleep(4) # Coloda o scrip em pausa enquanto a página carrega

    df_types = pd.read_html(browser.html)[0]
    df_types['Kdex'] = df_types['Kdex'].str[-3:]
    df_types = df_types.drop(['Ndex','MS','Pokémon'], axis=1)
    df_types = df_types.rename(columns={'Kdex': 'No.'})
    return df_types
```

```python
df_types = types()
df= pd.merge(df, df_types, on='No.')
```

```python
df
```

```python
def Description(df):
    List_pokemons = df['Name'].unique()
    List_pokemons = [valor.lower() for valor in List_pokemons]
    Descricao_pokemons = pd.DataFrame()
    for i in List_pokemons:
        try:
            browser = Browser('chrome', options=options) 
            url = 'https://www.heypikachu.com/'+i+'#pokedex'
            browser.goto(url)
            print(i)
            df1 = pd.read_html(browser.html)[5]
            df1[0] = df1[0].str.replace(':', '')
            df1 = df1.set_index(0)
            df1 = df1.transpose()
            df1 = df1.dropna(axis=1)
            df1["Name"] = i
            Descricao_pokemons = Descricao_pokemons.append(df1)
            browser.close()
        except:
            browser = Browser('chrome', options=options) 
            url = 'https://www.heypikachu.com/'+i+'#pokedex'
            browser.goto(url)
            print(i)
            df1 = pd.read_html(browser.html)[7]
            df1[0] = df1[0].str.replace(':', '')
            df1 = df1.set_index(0)
            df1 = df1.transpose()
            df1 = df1.dropna(axis=1)
            df1["Name"] = i
            Descricao_pokemons = Descricao_pokemons.append(df1)
            browser.close()
    return Descricao_pokemons
```

```python
Descricao_pokemons = Description(df)
```

```python
df['Name'] = df['Name'].str.lower()
df_merged = pd.merge(df, Descricao_pokemons, on='Name')
```
